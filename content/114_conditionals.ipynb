## Nesting

```python
print('i j')
for i in range(4):
    for j in range(3):
        print(i, j)
```

```python
print('i j')
for i in range(4):
    for j in range(3):
        if i != j:
            print(i,j)
```

```python
for i in x:
    if i < 4:
        print('{} is less than 4'.format(i))
    elif i > 4:
        print('{} is greater than 4'.format(i))
```

```python
for i in x:
    if i < 4:
        print('{} is less than 4'.format(i))
    elif i > 4:
        print('{} is greater than 4'.format(i))
    else:
        print('{} is equal to 4'.format(i))
```

```python
with open('data/airports.csv', 'r') as f:
    lines = f.readlines()
```

```python
data = []
for line in lines:
    data.append(line.strip().split(','))
```

```python
k = len(header)
for record in data:
    rec_k = len(record)
    if rec_k != k:
        print(rec_k, record)
```

```python
k = len(header)
for i, record in enumerate(data):
    rec_k = len(record)
    if rec_k != k:
        print(i, rec_k, record)
```
